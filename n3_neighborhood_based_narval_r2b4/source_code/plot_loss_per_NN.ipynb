```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_losses = []
with open('../saved_models/model_region_based_final_1.txt', 'r') as file:
    lines = file.readlines()
    for i in range(len(lines)):
        line = lines[i]
        if line.startswith('Test loss'):
            test_losses.append(float(line.split(' ')[3]))
```

```python
models = np.arange(6, 32, 1)

fig = plt.figure()
ax = fig.add_subplot(111, title='Region-based NN test set RMSE per model', ylabel='Model/Vertical layer', xlabel='Test set SER/RMSE')
plt.plot(np.sqrt(test_losses[1:]), models)
plt.gca().invert_yaxis()
plt.savefig('../figures/test_RMSE_per_model.pdf')
```
