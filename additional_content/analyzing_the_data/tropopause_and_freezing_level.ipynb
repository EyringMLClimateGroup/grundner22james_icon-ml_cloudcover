```python
import sys
import numpy as np
import matplotlib.pyplot as plt

# Add path with my_classes to sys.path
sys.path.insert(0, '/pf/b/b309170/workspace_icon-ml/cloud_cover_parameterization/')

from my_classes import load_data
```

```python
# NARVAL
order_of_vars = ['temp']
data_dict_narv = load_data(source='narval', days='all', vert_interp=True, order_of_vars=order_of_vars)
```

```python
data_dict_narv['temp'].shape
```

**Tropopause** <br>
Location of temperature inversion <br>
--> Layer 23/**24**

```python
a = np.arange(21, 48)
plt.plot(np.mean(data_dict_narv['temp'], axis=(0,2))[4:], a)
plt.grid(b=True)
plt.gca().invert_yaxis()
```

**Freezing level** <br>
Where T<0 on average <br>
--> Layer **34**

```python
a = np.arange(21, 48)
plt.plot(np.mean(np.array(data_dict_narv['temp']) - 273.15, axis=(0,2))[4:], a)
plt.grid(b=True)
plt.gca().invert_yaxis()
```
